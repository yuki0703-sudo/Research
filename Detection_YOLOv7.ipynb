```python
import detect3
import torch

with torch.no_grad():
  data = detect3.detect(source="/content/gdrive/MyDrive/YAMAKAWA_LAB/YOLO_labelimg/Images/13_files/exp_gray.mp4" ,
                        weights=['runs/train/custom_dataset_gray3/weights/last.pt'],
                        classes = None,agnostic = False,device='',save_img=False)
```

#Writing the ROI data in csv file

```python
import csv
#出力ファイル名
filename='/content/gdrive/My Drive/YAMAKAWA_LAB/YOLO_labelimg/Images/13_files/gray/gray.csv'

#ファイルを作成
with open(filename,"w") as f:
  writer = csv.writer(f)
  writer.writerow(["frame","x_TopLeft","y_TopLeft","x_BottomRight","y_BottomRight","Confidence","class"])  
```

```python
for i in range(len(data)): #frame
    for j in range(len(data[i])):#num of detected objects
      print((i,j))
      with open(filename,"a",newline='') as f:
        writer = csv.writer(f)
        writer.writerow([i+1,data[i][j][0],data[i][j][1],data[i][j][2],data[i][j][3],data[i][j][4],data[i][j][5]])
```
